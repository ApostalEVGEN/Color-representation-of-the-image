```python
# Загрузка библиотек

import numpy as np
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import matplotlib.pylab as plt
import cv2

im = imread('C:\\Users\\Apost\\Desktop\\aspid.jpg')
im1 = rgb2lab(im)
```

```python
im1[...,1] = im1[...,2] = 0
```

```python
im_grey = lab2rgb(im1)
```

```python
# Преобразование изображения в серые тона

plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'), plt.title('Original image', size=20)
plt.subplot(122), plt.imshow(im_grey), plt.axis('off'), plt.title('Gray scale image', size=20)
plt.show()
cv2.imwrite('output_image.jpg', im)
cv2.imwrite('output_image2.jpg', image)
```

```python
# Осветлим наше следующее изображение, с которым будем работать

img = cv2.imread('C:\\Users\\Apost\\Desktop\\new_zealand.jpg', 1)
#cv2.imshow("Original image",img)
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(img), plt.axis('off'), plt.title('Original image', size=20)
plt.show()
```

```python
# CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=30, tileGridSize=(8,8))
# О параметре clipLimit. Он определят, насколько контрастнее станет фото.
# Правда, мы не может бесконечно увеличивать этот параметр, есть определенный предел, при котором шумы испортят фоточку.
```

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # преобразование из цветового пространства BGR в цветовое пространство LAB
l, a, b = cv2.split(lab)  # разделение на 3 разных канала
```

```python
l2 = clahe.apply(l)  # нанесение CLAHE на L-канал
```

```python
lab = cv2.merge((l2,a,b))  # объединение каналов
img2 = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # преобразование из LAB в BGR
#cv2.imshow('Increased contrast', img2)
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(img2), plt.axis('off'), plt.title('Light image', size=20)
plt.show()
```
